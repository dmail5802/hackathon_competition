```python
(heart := pl.read_csv('../data/heart_disease/train.csv')
    .with_columns(
        pl.when(pl.col('Cholesterol') >= 240).then(1)
        .otherwise(0).alias('high_cholesterol'),
        pl.when(pl.col('Chest pain type') >= 3).then(1)
        .otherwise(0).alias('severe_chest_pain')
    )

)
```

```python
from sklearn.linear_model import LinearRegression
# Prepare the data
# Convert Heart Disease to binary (1 for Presence, 0 for Absence)
heart_data = heart.with_columns(
    pl.when(pl.col('Heart Disease') == 'Presence').then(1).otherwise(0).alias('Heart_Disease_Binary')
)

# Select predictors and target
predictors = ['Cholesterol', 'BP', 'Chest pain type', 'Sex', 'high_cholesterol', 'severe_chest_pain']
X = heart_data[predictors].to_numpy()
y = heart_data['Heart_Disease_Binary'].to_numpy()

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Display results
print("Linear Regression Model Results")
print("=" * 50)
print(f"Model Intercept: {model.intercept_:.6f}")
print("\nCoefficients:")
for predictor, coef in zip(predictors, model.coef_):
    print(f"  {predictor}: {coef:.6f}")

# Calculate R-squared
r_squared = model.score(X, y)
print(f"\nR-squared: {r_squared:.6f}")

# Show model predictions on first few samples
print("\nFirst 5 Predictions vs Actual:")
predictions = model.predict(X[:5])
for i, (pred, actual) in enumerate(zip(predictions, y[:5])):
    print(f"  Sample {i+1}: Predicted = {pred:.4f}, Actual = {actual}")
```
